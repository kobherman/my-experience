# Removing hieroglyphs

```python
all_tags_txt = open(new_tags_folder + "text_wout_utf.txt", 'w')
temp_txt = ''

for i in range(1, 25001):
    with open(all_tags_folder + f"tags{i}.txt", "r", encoding="utf-8") as f:
        temp_txt = f.read()[:-1]
    
    for word in temp_txt.split():
        try:
            if len(word) < 20:
                all_tags_txt.write(word)
                all_tags_txt.write(' ')
        except:
            print( word )
    all_tags_txt.write('\n')


all_tags_txt.close()
```
